```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x= 6* np.random.rand(100, 1)-3
y=0.5 * x**2+1.5*x+2+np.random.randn(100,1)
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg.predict(x_test))
print(score)
```

```python
plt.plot(x_train,reg.predict(x_train),color='red')
plt.scatter(x_train,y_train,color='blue')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.fit_transform(x_test)
```

```python
x_train_poly
```

```python
from sklearn.metrics import r2_score
reg=LinearRegression()
reg.fit(x_train_poly,y_train)
y_pred=reg.predict(x_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python

plt.scatter(x_train,reg.predict(x_train_poly),color='red')
plt.scatter(x_train,y_train)
plt.show()
```

```python
poly=PolynomialFeatures(degree=3,include_bias=True)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.fit_transform(x_test)
```

```python
x_train_poly
```

```python
x_new=np.linspace(-3,3,200).reshape(200,1)
x_new_poly=poly.transform(x_new)
x_new_poly
```

```python
# Re-train the model with the correct polynomial features before prediction
poly = PolynomialFeatures(degree=3, include_bias=True) # Ensure the degree matches x_new_poly
x_train_poly = poly.fit_transform(x_train)
reg.fit(x_train_poly, y_train)  # Re-fit the model

# Now you can predict:
y_new = reg.predict(x_new_poly)
plt.plot(x_new, y_new, "r-", color='red',label="New prediction")
plt.plot(x_train, y_train, "b.", color='blue',label="Training point")
plt.plot(x_test, y_test, "g.", color='green',label="Test point")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
